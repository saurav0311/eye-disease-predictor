```python
import tensorflow as tf
from tensorflow.keras.applications.mobilenet_v3 import preprocess_input
import numpy as np
import os
```

```python
model=tf.keras.models.load_model("Trained_Eye_disease_model.keras")
```

```python
model.summary()
```

```python
image_path="test/DRUSEN/DRUSEN-224974-19.jpeg"
img=tf.keras.utils.load_img(image_path,target_size=(224,224))
x=tf.keras.utils.img_to_array(img)
x=np.expand_dims(x,axis=0)
x=preprocess_input(x)
```

```python
x.shape
```

```python
pred=model.predict(x)
```

```python
pred
```

```python
class_name=['CNV','DME','DRUSEN','NORMAL']
result_index=np.argmax(pred)
print("The predicted image is:",class_name[result_index])
```
